Conclusions regarding the data so far:
1) There is no missing data
2) Columns 0, 10-377 are all integers, Column 1 is float Columns 2,3,4,5,6,7,8,9 are all categorical variables and hence need Label Encoding followed by OneHotEncoding
3) Columns ['X11', 'X93', 'X107', 'X233', 'X235', 'X268', 'X289', 'X290', 'X293', 'X297', 'X330', 'X347'] are all zero values so these columns can be dropped from the analysis.

```python
#Prepare numpy arrays and drop the irrelevent columns
y = training_data['y'].values
x = training_data.drop(["ID", "y", 'X11', 'X93', 'X107', 'X233', 'X235', 'X268', 'X289', 'X290', 'X293', 'X297', 'X330', 'X347'], axis=1)
```

```python
#Encode string labels into integers
for cols in ["X0", "X1", "X2", "X3", "X4", "X5", "X6", "X8"]:
    label_encoder = LabelEncoder()
    label_encoder.fit(x[cols].values)
    x[cols] = label_encoder.transform(x[cols].values)
#Examine the feature vectors    
print("x shape ",x.shape)
x.head()
```

```python
#Do OneHotEncoding and keep adding the columns to a new feature vector X
for i in range(8):
    temp = x.iloc[:,i:i+1].values
    onehotencoder = OneHotEncoder(categorical_features=[0])
    temp = onehotencoder.fit_transform(temp).toarray()
    if i==0:
        X=temp[:,1:]
    else:
        X=np.column_stack((temp[:,1:],X))
X=np.column_stack((x.iloc[:,8:364],X))
```

```python
# Split into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

Regression

```python
# Create linear regression object
regr = linear_model.LinearRegression()
# Train the model using the training sets
regr.fit(X_train, y_train)
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % np.mean((regr.predict(X_test) - y_test) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(X_test, y_test))
```

Artificial Neural Network

```python
m,n=X_train.shape
ANN_classifier = Sequential()
ANN_classifier.add(Dense(units = 200, kernel_initializer = 'uniform', activation = 'relu', input_dim = n))
ANN_classifier.add(Dense(units = 200, kernel_initializer = 'uniform', activation = 'relu'))
ANN_classifier.add(Dense(units = 1, kernel_initializer = 'uniform'))
ANN_classifier.compile(optimizer = 'adam', loss='mean_squared_error', metrics = ['accuracy'])
history = ANN_classifier.fit(X_train, y_train, batch_size = 50, epochs = 100)
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.ylim([0,100])
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
y_pred = ANN_classifier.predict(X_test)
plt.scatter(y_test,y_pred)
plt.xlim([0,150])
plt.ylim([0,150])
plt.show()
```
